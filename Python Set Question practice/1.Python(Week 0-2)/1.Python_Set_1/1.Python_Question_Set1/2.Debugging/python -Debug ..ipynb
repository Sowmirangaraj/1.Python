```python
#2.What will be the output of this code?

a = 10
b = "5"
print(a + b)
 
```

```python
#2.Corrected code: variable name with datatype to take as number

a = 10
b = "5"
print(a + int(b))
 
```

```python
#3.What is the issue in this Python snippet?

 def add_numbers(a, b)
    return a + b
```

```python
#3Answer: Missing colon : after (a, b)

def add_numbers(a, b):
    return a + b
```
